 5.1 - Solve the first order differential equation dx/dt+ 2x = 0 with the initial condition x(0) = 1

```
 
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
# function that returns dx/dt
def model(x,t):
  dxdt = -2 * x
  return dxdt
# initial condition
x0 = 1
# time points
t = np.linspace(0,10)
# solve ODE
y = odeint(model,x0,t)
# plot results
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.title('Solution for first order ODE')
plt.show()
```

5.3 - Solve the second order differential equation dx/dt + 2dx/dt + 2x = e ^-t

```
 
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def dx_dt(U, t):
 # Here are the eqns
  return [U[1], -2*U[1] - 2*U[0] + math.exp(-t)]
x0 = [0, 0] #inital conditions x[0] & x'[0]
ts = np.linspace(0, 10, 100) #time points from 0 to 10
# solving equations using ode
Us = odeint(dx_dt, x0, ts)
xs = Us[:,0]
# plotting result
plt.xlabel("t")
plt.ylabel("x")
plt.title("Solution for second order ODE")
plt.plot(ts,xs)
plt.show()
```

5.4 - Solve the current transient through a series RLC circuit with R = 10, L = 1 mH and C = 1 uF that is driven by (a) 5 V DC (b) the signal 5e^-t U(t)

```
 
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
fig,a=plt.subplots(2,1)
V = 5.0 #voltageSource
R = 1.0
L=1.0e-3 #1mH
C = 1.0e-6 #1microF
I=V/R # inital current
 
def model(i,t):
    return[L*i[1], -R*i[1] - (1/C)*i[0]]
x0 = [I, 0] #inital conditions 
ts = np.linspace(0, 15, 100) 
#time points from 0 to 10
Us = odeint(model, x0, ts)
xs = Us[:,0]
a[0].set_xlabel("t")
a[0].set_ylabel("i")
a[0].set_title("Solution for current in RLC series circuit for 5 V DC")
a[0].plot(ts,xs)
def model1(i,t):
    return[L*i[1], -R*i[1] - (1/C)*i[0] -5*math.exp(-t)]
 
ts1 = np.linspace(0, 15, 100) #time points from 0 to 10
Us = odeint(model, x0, ts1)
xs1 = Us[:,0]
a[1].set_xlabel("t")
a[1].set_ylabel("i")
a[1].set_title("Solution for transient current at 5exp(-t) u(t)")
a[1].plot(ts1,xs1);
plt.subplots_adjust(wspace=0.9,hspace=0.5,left=0.05,right=1.5,top=2,bottom=0.1)
plt.show()
```
