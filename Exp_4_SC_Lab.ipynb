4.2 - Compute the first and second derivatives of above functions using built in tool such as gradient and plot the derivatives over the respective functions and appreciate.

```
 
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*np.pi,180)
y = np.sin(x)
derivy = np.gradient(y,(2*np.pi/180))
der2y = np. gradient(derivy,(2*np.pi/180))
z= np.cos(x)
derivz = np.gradient(z,(2*np.pi/180))
der2z = np. gradient(derivz,(2*np.pi/180))
p=np.cosh(x)
derivp = np.gradient(p,(2*np.pi/180))
der2p = np. gradient(derivp,(2*np.pi/180))
q=np.sinh(x)
derivq = np.gradient(q,(2*np.pi/180))
der2q = np. gradient(derivq,(2*np.pi/180))
fig,a = plt.subplots(4,1)
a[0].plot(x,y,'g',label='Original function')
a[0].plot(x,derivy,'r',label='first derivative')
a[0].plot(x,der2y,'b',label='second derivative')
a[0].legend(loc='upper left')
a[0].set_title('Sin x')
a[1].plot(x,z,'g',label='Original function')
a[1].plot(x,derivz,'r',label='first derivative')
a[1].plot(x,der2z,'b',label='second derivative')
a[1].legend(loc='upper left')
a[1].set_title('Cos x')
a[2].plot(x,p,'g',label='Original function')
a[2].plot(x,derivp,'r',label='first derivative')
a[2].plot(x,der2p,'b',label='second derivative')
a[2].legend(loc='upper left')
a[2].set_title('Sin hx')
a[3].plot(x,q,'g',label='Original function')
a[3].plot(x,derivq,'r',label='first derivative')
a[3].plot(x,der2q,'b',label='second derivative')
a[3].legend(loc='upper left')
a[3].set_title('Cos hx')
plt.subplots_adjust(wspace=0.9,hspace=0.5,left=0.05,right=1.5,top=2,bottom=0.1)
plt.show()
```
